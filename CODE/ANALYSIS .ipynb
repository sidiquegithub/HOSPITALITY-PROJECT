```python
# Total revenue gerated in each week and total revenue realised
```

```python
rev_gen = df_2.groupby('week no')['revenue_generated'].sum()
rev_gen 
```

```python
rev_real = df_2.groupby('week no')['revenue_realized'].sum()
rev_real 
```

```python
rev_dif = rev_gen -  rev_real 
rev_dif
```

```python
# Plotting
plt.figure(figsize=(8, 4))

# Plot 'rev_gen'
plt.plot(rev_gen.index, rev_gen, label='Revenue Generated')

# Plot 'rev_real'
plt.plot(rev_real.index, rev_real, label='Revenue Realised')

# Plot 'rev_dif'
plt.plot(rev_dif.index, rev_dif, label='Revenue Difference')

# Adding labels and title
plt.xlabel('Week Number')
plt.ylabel('Revenue')
plt.title('Revenue Comparison by Week')
plt.legend()

# Show the plot
plt.show()
```

```python
# RESULT 
# High revenue on the weeks 19,20,22,24,25,27,28,29. [revenue generated > 15 cr] and [revenue real > 13 CR]
# This shows week 19 to 23 is for May, Week 23 to 27 to June and Week 27 to 32 for July.  
```
