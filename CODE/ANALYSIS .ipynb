```python
rev_gen = df_2.groupby('week no')['revenue_generated'].sum()
rev_gen 
```

```python
rev_real = df_2.groupby('week no')['revenue_realized'].sum()
rev_real 
```

```python
rev_dif = rev_gen -  rev_real 
rev_dif
```

```python
# Plotting
plt.figure(figsize=(8, 4))

# Plot 'rev_gen'
plt.plot(rev_gen.index, rev_gen, label='Revenue Generated')

# Plot 'rev_real'
plt.plot(rev_real.index, rev_real, label='Revenue Realised')

# Plot 'rev_dif'
plt.plot(rev_dif.index, rev_dif, label='Revenue Difference')

# Adding labels and title
plt.xlabel('Week Number')
plt.ylabel('Revenue')
plt.title('Revenue Comparison by Week')
plt.legend()

# Show the plot
plt.show()
```
